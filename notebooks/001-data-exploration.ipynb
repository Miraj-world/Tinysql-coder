# 001 - Data Exploration

Goal: load the local BIRD mini-dev Arrow dataset and inspect the first few examples.

This dataset is not bird images. **BIRD** here is a text-to-SQL benchmark dataset. Each row contains a natural-language question, optional evidence, the target SQL query, and a difficulty label.

```python
from pathlib import Path

import pandas as pd
from datasets import Dataset

pd.set_option("display.max_colwidth", 140)
```

```python
PROJECT_ROOT = Path.cwd()

# When this notebook is opened from the notebooks folder, move up to the project root.
if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent

DATASET_PATH = PROJECT_ROOT / "data" / "bird_mini_dev" / "raw" / "bird-original.arrow"

print(f"Project root: {PROJECT_ROOT}")
print(f"Dataset path: {DATASET_PATH}")
print(f"Dataset exists: {DATASET_PATH.exists()}")
```

```python
ds = Dataset.from_file(str(DATASET_PATH))

print(ds)
print("Columns:", ds.column_names)
```

```python
first_10 = ds.select(range(10)).to_pandas()
first_10
```

```python
first_10[["question_id", "difficulty", "question", "SQL"]]
```
